5. Drop all rows with N/A,NA,na values in Num_Bedrooms

```python
df2 = df1.copy() 
df2.drop(df1[df1['NUM_BEDROOMS'].isna()].index, inplace=True)
df2.drop(df1[df1.NUM_BEDROOMS  == 'na'].index, inplace=True)
df2
```

6. Replace the NaN and String entries in Num_Bath with previous entry

```python
df3 = df1.copy()
df3['NUM_BATH'] = pd.to_numeric(df3['NUM_BATH'], errors='coerce')
df3['NUM_BATH'] = df3['NUM_BATH'].fillna(method='pad')
df3
```

7. Replace the empty values in “SQ_FT” with the mean of the all entries.

```python
df3['SQ_FT'] = pd.to_numeric(df3['SQ_FT'], errors='coerce')
df3['SQ_FT'] = df3['SQ_FT'].fillna(df3['SQ_FT'].mean())
df3
```
